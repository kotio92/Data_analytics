## Matplotlib

Matplotlib è una libreria per la creazione di grafici per Python.

```python
import math
import numpy as np
import pandas as pd
```

pyplot è un sottomodulo di matplotlib che contiene funzioni che consentono a matplotlib di funzionare come MATLAB.

```python
import matplotlib.pyplot as plt
```

Ogni funzione di pyplot genera una diversa figura; ad esempio c'è una funzione che crea una figura, una che vi genera un grafico 
dentro, una che genera funzione, una che ci mette i titoli sugli assi ecc.

### Tipi di grafici

- Bar: Make a bar plot.
- Barh: Make a horizontal bar plot.
- Boxplot: Make a box and whisker plot.
- Hist: Plot a histogram.
- hist2d: Make a 2D histogram plot.
- Pie: Plot a pie chart.
- Plot: Plot lines and/or markers to the Axes.
- Polar: Make a polar plot.
- Scatter: Make a scatter plot of x vs y.
- Stackplot: Draws a stacked area plot.
- Stem: Create a stem plot.
- Step: Make a step plot.
- Quiver: Plot a 2-D field of arrows

### Funzioni sugli assi

- Axes: Add axes to the figure.
- Text: Add text to the axes.
- Title: Set a title of the current axes.
- Xlabel: Set the x axis label of the current axis.
- Xlim: Get or set the x limits of the current axes.
- Xscale: Set the scaling of the x-axis.
- Xticks: Get or set the x-limits of the current tick locations and labels.
- Ylabel: Set the y axis label of the current axis.
- Ylim: Get or set the y-limits of the current axes.
- Yscale: Set the scaling of the y-axis.
- Yticks: Get or set the y-limits of the current tick locations and labels.

### Funzioni sulle figure

- Figtext: Add text to figure.
- Figure: Creates a new figure.
- Show: Display a figure.
- Savefig: Save the current figure.
- Close: Close a figure window.

### Un primo semplice grafico

```python
# Valori che vanno da 0 a 2pigreco con passo = 0.05
x = np.arange(0,2*math.pi,0.05)
print(type(x))
x
```

```python
y = np.sin(x)
print(type(y))
```

```python
plt.plot(x,y)                  # plot() è una funzione del sottomodulo pyplot di matplotlib che ritorna un grafico
                               # Vanno indicati gli array per ascisse e ordinate
```

Rinominiamo gli assi e diamo un titolo.

```python
plt.xlabel('angolo')            # rinomino asse x
plt.ylabel('seno')              # rinomino asse y
plt.title('Funzione Seno')      # scrivo titolo
plt.plot(x,y)                   # traccio il grafico
plt.show()                      # funzione che invoca la viewer window
```

### Interfaccia Object-oriented

Sebbene sia vero che è semplice generare grafici col modulo matplotlib.pyplot, è anche vero che tramitte l'approccio
object oriented tipica della programmazione è possibile customizzare ulteriormente i grafici.

La maggior parte delle funzioni sono nella classe matplotlib.axes.Axes

L'idea principale è quella di creare, attraverso la programmazione object oriented, oggetti 'figure' e poi andare a 
richiamare suoi metodi. 

In quest'ottica, Pyplot viene utilizzato solo per poche funzioni come quella che istanzia una 'figure'; dunque
l'utente dovrà preoccuparsi di generare semplicemente un oggetto 'figure' e un oggetto 'axis', dei quali dovrà
tenere traccia. Perciò prima viene creato un oggetto 'figure', poi, tramite quest'oggetto, vengono creati uno 
o più oggetti 'axis', dei quali si andranno a richiamare i metodi.

```python
# Innanzitutto creo una figura vuota

fig = plt.figure()
```

```python
# Ora utilizzo un metodo (add_axes) dell'oggetto creato (figure) per aggiungere gli assi alla figura
fig = plt.figure()                   # istanzio una figura (vuota)

ax = fig.add_axes([0,0,1,1])         # settaggio del riguadro che ospiterà il grafico
# [left,bottom, width, height]

print(type(ax))   # è un oggetto 'Axes'

ax.plot(x,y)                         # plotto il grafico che viene fuori impostando 'x' come ascissa e 'y' come ordinata
ax.set_title('Funzione seno')        # Metodo che mi fa scrivere titolo del grafico
ax.set_xlabel('x')                   # Metodo che mi fa scrivere il nome delle ascisse
ax.set_ylabel('y')                   # Metodo che mi fa scrivere il nome delle ordinate
plt.show()                           # mostra la figura (non più vuota)                     
```

#### Pyplot.figure

Il modulo matplotlib.figure contiene la classe **Figure**: è un contenitore di livello superiore per tutti gli elementi grafici. 

```python
# L'oggetto Figure viene istanziato chiamando la funzione figure () da modulo pyplot

fig = plt.figure()
```

Vediamo i parametri addizionali:
- Figsize: (width,height) tuple in inches
- Dpi: Dots per inches
- Facecolor: Figure patch facecolor
- Edgecolor: Figure patch edge color
- Linewidth: Edge line width

L'oggetto 'Axes' è la regione dell'immagine con i dati. 

Una istanza della classe 'figure' può contenere più oggetti '**Axes**', ma un oggetto 'Axes' può essere in una sola 'figure'.

L'oggetto 'Axes' può contenere 2 (o 3) oggetti '**Axis**'. 

L'oggetto 'Axes' viene aggiunto a un'istanza di 'figure' invocando il metodo add_axes(), il quale ritorna un oggetto 'Axes', il quale aggiunge un riquadro predisposto a ospitare un grafico alla posizione [left, bottom, width, height] dove tutte le quantità sono espresse in percentuale di altezza e lunghezza della 'figure'.

```python
# Esempio
fig = plt.figure()

ax1 = fig.add_axes([1,1,1,1])
ax2 = fig.add_axes([0,0,0.8,1])

ax1.plot(x,y,'r+', label='Grafico ax1')
ax1.legend(loc=3, shadow=True, fontsize='x-large')

ax2.plot(x,y,'c--', label='Grafico ax2')
ax2.legend(loc=2, shadow=True, fontsize='medium')  # descritto più sotto

fig.show()
```

#### Metodo *legend*

Si tratta di un metodo della classe 'Axes' che mi consente di aggiungere la legenda a un grafico. Riporta la 'label' specificata nel grafico.

Vuole 3 parametri di base:
* loc : posizione della legenda nel grafico (intero da 0 a 10)
* shadow : ombreggitura legenda
* fontsize : font

#### loc

* Best : 0
* upper right : 1
* upper left : 2
* lower left : 3
* lower right : 4
* Right : 5
* center left : 6
* Center right : 7
* lower center : 8
* upper center : 9
* Center : 10

Vediamo ora un esempio di tracciamento di due grafici sulla stessa figure:

```python
y = np.array([4,9,16,25,36,49,64,81,100])
x1 = np.array([2,3,4,5,6,7,8,9,10])
x2 = np.array([1,3,5,7,9,12,15,18,20])

# Istanzio figura
figura = plt.figure()

# Istanzio primo grafico
f = figura.add_axes([1,0,1,1])
print(type(f))
f.plot(x1,y,'r-', label='g1')

# Istanzio secondo grafico
print(type(f))
f.plot(x2,y,'g-.', label='g2')
f.legend(labels=('g1','g2'), loc=2, shadow = True, fontsize='large')

f.set_title('Grafici')
f.set_xlabel('Ascisse')
f.set_ylabel('Ordinate')

figura.show()
```

## Bar plot

Il metodo *bar* della classe 'Axes' genera un bar chart dotato delle seguenti caratteristiche

#### ax.bar(x, height, width, bottom, align)

* x : sequence of scalars representing the x coordinates of the bars. align controls if x is the bar center (default) or left edge.
* height : scalar or sequence of scalars representing the height(s) of the bars
* width : scalar or array-like, optional. the width(s) of the bars default 0.8
* bottom : scalar or array-like, optional. the y coordinate(s) of the bars default None
* align : {‘center’, ‘edge’}, optional, default ‘center’


```python
# I parametri basilari sono x e height, ovvero i primi 2

x = ['C++','Java','Python','Scala','C#']  # x
y = [10,25,27,6,12]                       # height

# Istanzio figura
fg = plt.figure()

# Istanzio oggetto 'Axes'
a = fg.add_axes([1,1,1,1])

a.bar(x,y)                # uso il metodo bar() della classe 'Axes'
a.set_title('Corsi di programmazione')
a.set_ylabel('Numero studenti')
a.set_xlabel('Corso')

fg.show()
```

### Pie Chart

Vediamo lo stesso esercizio ma in un grafico a torta.

```python
x = ['C++','Java','Python','Scala','C#']  # x
y = [10,25,27,6,12]                       # valori assoluti

# Istanzio figura
fg = plt.figure()

# Istanzio oggetto 'Axes'
a = fg.add_axes([1,1,1,1])

a.pie(y, labels = x, autopct='%1.2f%%')                # autopct è il parametro che mi permette di esprimere i valori
                                                       # assoluti o relativi delle singole fette
a.set_title('Studenti per ciascun corso di programmazione')

fg.show()
```

### Multiple bars

Vediamo come si setta un grafico con più barre

```python
fg = plt.figure()

ax = fg.add_axes([1,1,1,1])

# Scrivo i dati: profitto nei anni di 3 società

data = [[25,30,31,28],[33,35,35,38],[22,25,28,29]]

# Scrivo un array di ascisse di comodo (che poi rinominerò con gli anni)

x = np.arange(4)

# vado a creare i bars andado a utilizzare come ascissa un array di comodo che modificheremo con un ampiezza costante
# (width) e aumentando le ascisse in maniera tale da tenere le barre 'azzeccate'
ax.bar(x ,data[0], color = 'r',width= 0.20,  label='pwc')
ax.bar(x + 0.20 ,data[1], color='b',  width=0.20,   label='kpmg')
ax.bar(x + 0.40 ,data[2], color = 'g',width = 0.20, label ='deloitte')
fg.show()

# Ora vado a ridisegnare l'asse delle ascisse evidenziando solo dei punti particolari
ax.set_xticks([0.20,1.20,2.20,3.20])

# Ora sostituisco le ascisse evidenziato con gli anni
ax.set_xticklabels(['2015','2016','2017','2018'])

# Inserisco la legenda
ax.legend(labels=('pwc','kpmg','deloitte'),loc=2, shadow=True, fontsize='medium')
```

### Stacked bars

Vediamo come ottenere un grafico a stacked bars

```python
fg = plt.figure()

ax = fg.add_axes([1,1,1,1])

# Immaginiamo di avere una classe di liceo il cui num di maschi e femmine varia in 4 anni

boys = [12,13,10,7]
girls = [15,18,20,18]
x = np.arange(4)

ax.bar(x,boys,color = 'b',width=0.35)                    # maschi
ax.bar(x,girls, color='r', bottom = boys, width=0.35)    # indicando come parametro 'bottom' proprio il valore in boys,
                                                         # ci troviamo che il bar parte dal numero di boys
    
ax.legend(labels=('boys','girls'), loc=2, shadow=True, fontsize= 'medium')
ax.set_xticks([0,1,2,3])
ax.set_xticklabels([2015,2016,2017,2018])
fig.show()
```
